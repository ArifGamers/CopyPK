```python
nama = input("Nama = ")
print(f'Halo {nama}, selamat datang. Masukkan persamaan kuadratmu!')

import math
import cmath
import re
from fractions import Fraction

print('')
print('Bentuk umum persamaan kuadrat: f(x) = ax^2 + bx + c = 0')
print('(Jika koefisien berbentuk pecahan, tulis a/b, gunakan titik jika berbentuk desimal)')

print('')
a_str = input("Nilai a = ")
b_str = input("Nilai b = ")
c_str = input("Nilai c = ")

def convert_to_number(s):
    try:
        return int(s)
    except ValueError:
        try:
            return float(s)
        except ValueError:
            return Fraction(s)

a = convert_to_number(a_str)
b = convert_to_number(b_str)
c = convert_to_number(c_str)

print('')
def build_equation(a, b, c):
    terms = []
    if a:
        if isinstance(a, int):
            terms.append(f"-x^2" if a == -1 else f"{a}x^2")
        elif isinstance(a, Fraction):
            terms.append(f"{a}x^2")
        else:
            terms.append(f"{a}x^2")
    if b:
        sign = "-" if b < 0 else "+"
        if isinstance(b, int):
            terms.append(f"{sign} x" if abs(b) == 1 else f"{sign} {abs(b)}x")
        elif isinstance(b, Fraction):
            terms.append(f"{sign} {b}x")
        else:
            terms.append(f"{sign} {b}x")
    if c:
        sign = "-" if c < 0 else "+"
        if isinstance(c, int):
            terms.append(f"{sign} {abs(c)}" if abs(c) != 1 else f"{sign} 1")
        elif isinstance(c, Fraction):
            terms.append(f"{sign} {c}")
        else:
            terms.append(f"{sign} {c}")
    return " ".join(terms).lstrip('+ ')

equation_str = build_equation(a, b, c)
print('Persamaanmu adalah: f(x) =', equation_str)

def convert_to_standard(a, b, c):
    if not (isinstance(a, int) and isinstance(b, int) and isinstance(c, int)):
        lcm = 1
        if not isinstance(a, int): lcm = math.lcm(lcm, int(1 / (a - int(a))))
        if not isinstance(b, int): lcm = math.lcm(lcm, int(1 / (b - int(b))))
        if not isinstance(c, int): lcm = math.lcm(lcm, int(1 / (c - int(c))))
        a_new = int(a * lcm)
        b_new = int(b * lcm)
        c_new = int(c * lcm)
        return a_new, b_new, c_new
    else:
        return int(a), int(b), int(c)
a_new, b_new, c_new = convert_to_standard(a, b, c)

standard_form = ""
if a_new != 0:
    standard_form += f"{a_new}x^2"
if b_new != 0:
    if standard_form:
        standard_form += f" {'-' if b_new < 0 else '+'} {abs(b_new)}x"
    else:
        standard_form += f"{b_new}x"
if c_new != 0:
    if standard_form:
        standard_form += f" {'-' if c_new < 0 else '+'} {abs(c_new)}"
    else:
        standard_form += f"{c_new}"
a_new, b_new, c_new = convert_to_standard(a, b, c)

print(f"atau dalam bentuk umumnya (kedua ruas dikali KPK): f(x) = {standard_form}")

print('')
D = b**2-4*a*c
print('Diskriminan =', D if D != int(D) else int(D))
if D<0:
    x1 = (-b+cmath.sqrt(D))/(2*a)
    x2 = (-b-cmath.sqrt(D))/(2*a)
    print('Memiliki 2 akar imajiner yang berbeda')
elif D==0:
    x1 = x2 = (-b/(2*a))
    print('Memiliki 2 akar real yang kembar')
elif D>0:
    x1 = (-b+math.sqrt(D))/(2*a)
    x2 = (-b-math.sqrt(D))/(2*a)
    print('Memiliki 2 akar real yang berbeda')

print('')
print('Akar f(x) = 0:')
if D<0:
    if x1.real == 0:
        if x1.imag == int(x1.imag):
            x1_str = f"{int(x1.imag)}i"
        else:
            x1_str = f"{x1.imag:.2f}i"
    else:
        if x1.real == int(x1.real):
            x1_real = f"{int(x1.real)}"
        else:
            x1_real = "{:.2f}".format(x1.real).rstrip('0').rstrip('.')
        if x1.imag == int(x1.imag):
            x1_imag = f"{int(abs(x1.imag))}"
        else:
            x1_imag = "{:.2f}".format(abs(x1.imag)).rstrip('0').rstrip('.')
        x1_sign = '+' if x1.imag >= 0 else '-'
        x1_str = f"{x1_real} {x1_sign} {x1_imag}i"
    if x2.real == 0:
        if x2.imag == int(x2.imag):
            x2_str = f"{int(x2.imag)}i"
        else:
            x2_str = f"{x2.imag:.2f}i"
    else:
        if x2.real == int(x2.real):
            x2_real = f"{int(x2.real)}"
        else:
            x2_real = "{:.2f}".format(x2.real).rstrip('0').rstrip('.')
        if x2.imag == int(x2.imag):
            x2_imag = f"{int(abs(x2.imag))}"
        else:
            x2_imag = "{:.2f}".format(abs(x2.imag)).rstrip('0').rstrip('.')
        x2_sign = '+' if x2.imag >= 0 else '-'
        x2_str = f"{x2_real} {x2_sign} {x2_imag}i"
    print('x1 =', x1_str)
    print('x2 =', x2_str)
elif D==0:
    x = (-b/(2*a))
    x_str = f"{int(x)}" if x == int(x) else f"{x:.2f}".rstrip('0').rstrip('.')
    print('x =', x_str)
elif D>0:
    x1_str = f"{int(x1)}" if x1 == int(x1) else f"{x1:.2f}".rstrip('0').rstrip('.')
    x2_str = f"{int(x2)}" if x2 == int(x2) else f"{x2:.2f}".rstrip('0').rstrip('.')
    print('x1 =', x1_str)
    print('x2 =', x2_str)
```
